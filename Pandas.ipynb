**Filtering Series**

```python
my_series = pd.Series(
    [0, 1, 2, 3, 4], index=["Day 0", "Day 1","Day 2", "Day 3", "Day 4"]
)
my_series
```

```python
my_series.loc[my_series != 2]
```

```python
my_series.loc[my_series.isin([1,2])]
```

```python
my_series.loc[~my_series.isin([1,2])]
```

```python
my_series.loc[my_series > 2]
```

**The Apply Method**

```python
import pandas as pd
import numpy as np
```

```python
string_series = pd.Series(['day 0', 'day 0', 'day 2', 'day 2', 'day 4'])
string_series
```

```python
string_series.apply(lambda x:x[-1])
```

```python
string_series.str[-1]
```

```python
def search(string, looking_for):
  if looking_for is string:
      return "Found It!"
  return "Nope!"
```

```python
a = string_series.str[-1]
a.apply(search, args = '2')
```
