**Grouping Dataframes**

```python
stores.groupby("state")
```

```python
stores.groupby("state")[['cluster']].mean()
```

```python
stores.groupby("state")[['cluster']].mean().sort_values('cluster', ascending = False)
```

**Multi-Index Dataframes**

```python
import pandas as pd
product = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/product.csv")
```

```python
product.head()
```

```python
product.groupby(["PRODUCT_ID", "BRAND"])[["MANUFACTURER"]].sum()
```

```python
agg_product = product.groupby(["PRODUCT_ID", "BRAND"])[["MANUFACTURER"]].sum()
```

```python
agg_product.head()
```

```python
agg_product.xs('National', level='BRAND')
```
