**Filtering Arrays**

```python
import numpy as np
my_array = np.arange(20)
my_array
```

```python
my_array % 2 == 0
```

```python
my_array[my_array % 2 ==0]
```

```python
even_odd = np.array(['even', 'odd']*10)
even_odd
```

```python
even_odd != 'odd'
```

```python
even_odd[even_odd != 'odd']
```

```python
my_array[even_odd != 'odd'] = 0
my_array
```

```python
mask = (even_odd != 'odd') | (even_odd != 'even')
even_odd[mask]
```

```python
mask = (even_odd != 'odd') & (even_odd != 'even')
even_odd[mask]
```

**Array Aggregation Methods**

```python
import numpy as np
rng = np.random.default_rng(616)
price = (rng.random(10) * 10).round(2)
price
```

```python
inventory = rng.integers(0, 100, 10)
inventory
```

```python
price.mean()
```

```python
inventory.mean()
```

```python
inventory.min()
```

```python
inventory.std()
```

```python
(price * inventory)
```

```python
(price * inventory).sum()
```

```python
price_2d = price.reshape(5, 2)
price_2d
```

```python
#sum of individual columns
price_2d.sum(axis = 0)
```

```python
# sum of rows
price_2d.sum(axis = 1)
```

```python
price_2d.mean(axis = 0)
```

**Array Functions**

```python
import numpy as np
price = np.array([2.6 , 7.2, 6.92, 3.88, 0.78, 2.67, 8.87, 0.23, 5.23, 6.3])
inventory = np.array([79, 22, 20, 48, 8, 76, 12, 16, 5, 64])
product_value = price * inventory
product_value
```

```python
product_value.sort()
product_value
```

```python
reshaped_product_value = product_value.reshape(2, 5)
reshaped_product_value
```

```python
np.median(reshaped_product_value)
```

```python
np.percentile(reshaped_product_value, 50)
```

```python
np.percentile(reshaped_product_value, 100)
```

```python
np.percentile(reshaped_product_value, 0)
```

```python
np.sqrt(reshaped_product_value)
```

```python
np.unique(reshaped_product_value)
```
